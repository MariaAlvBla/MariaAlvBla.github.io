```python
import pandas as pd
import plotly
import plotly.express as px

source_url = 'https://raw.githubusercontent.com/MariaAlvBla/MariaAlvBla.github.io/main/Dataset_Datathon.csv'

data_frame=pd.read_csv(source_url, sep=",",encoding="latin-1")

color_discrete_sequence = ['#35B779', '#440154', '#31688E', '#FDE725']

fig=px.sunburst(
    data_frame=data_frame,
    path=["env_broad_scale", "env_local_scale", "BioProject", "BioSample"],
    maxdepth=2,
    width=800,
    height=800,
    color_discrete_sequence=color_discrete_sequence)

fig.update_traces(insidetextorientation="radial",hovertemplate=None)
fig.show()
plotly.offline.plot(fig, filename='index.html')
```
